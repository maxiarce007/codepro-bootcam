```python
class coche:
    tipo='vehiculo'
    def __init__(self, modelo, posicion=0):
        self.modelo=modelo
        self.posicion=posicion

    def acelerar(self, metros):
        self.posicion += metros
        print(f'el vehiculo modelo {self.modelo} acelera y avanza {metros} m. nueva posicion {self.posicion}')
    def girar(self, direccion):
        print(f'el {self.modelo} avanza y gira a la direccioon {direccion}')

coche1=coche("toyota")
coche1.posicion
coche1.acelerar(1)
coche1.girar("izquierda")
```
